```python
bedrock.list_foundation_models()
```

```python
question = 'What happens to time near a Black Hole?'
```

```python
bedrock_runtime = boto3.client(region_name='us-east-1', service_name='bedrock-runtime')
```

```python
import json
```

```python
body = json.dumps({'inputText': question}) 
```

Again, you'll need to give access to this model inside AWS; giving access to the account gives access to all the users in there.

```python
response = bedrock_runtime.invoke_model(body=body, modelId='amazon.titan-text-express-v1')
```

```python
response_body = json.loads(response.get('body').read())
```

```python
response_body
```
